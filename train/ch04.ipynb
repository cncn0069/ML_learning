Q1 X,y 학습데이터와 정답지를 나눔

```python
#X
X = df.drop("income", axis=1)
y = df["income"].apply(lambda x: 1 if x.strip() == "<=50K" else 0)
y.value_counts()
```

Q Income을 예측하는 간단한 모델을 나눔

```python
X_train, X_test, y_train,y_test = train_test_split(X,y,stratify=y)
```

Q3 수치형 데이터와 범주형 데이터를 분류하시오

```python
df.info()
```

```python
numeric_feat = ["age","fnlwgt","education-num","capital-gain","capital-loss","hours-per-week"]
categorical_feat = ["workclass","education","marital-status","occupation","relationship","race","gender","native-country","income"]
```

```python
#수치형

numeric_transformer = SimpleImputer(strategy="median")
X_train_numeric = numeric_transformer.fit_transform(X_train[numeric_feat])
X_test_numeric = numeric_transformer.transform(X_test[numeric_feat])
```

```python
#범주형
encoder = OneHotEncoder(handle_unknown="ignore")
X_train_processe = encoder.fit_transform(X_train[categorical_feat])
X_test_processed = encoder.transform(X_test)
```

```python
from sklearn.datasets import load_iris
iris = load_iris()

df = pd.DataFrame(iris.data)
df
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(random_state=42)
model.fit(X_train_processe, y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = model.predict(X_test_processed)
accuracy_score(y_test, y_pred)
```
